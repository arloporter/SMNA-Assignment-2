```python
import cudf
import cugraph

gdf = cudf.read_csv('/content/drive/MyDrive/data/network_fixed.csv')
gdf_clean = gdf.dropna()
G = cugraph.Graph()
G.from_cudf_edgelist(gdf_clean, source="Source", destination="Target", edge_attr='Weight')
```

```python
bc = cugraph.betweenness_centrality(G, 26000)
print(bc)
bc.to_csv('/content/drive/MyDrive/data/annex_betweenness_centrality.csv')
```

```python
sorted = bc.sort_values("annex_betweenness_centrality", ascending=False)
sorted.head()
```

```python
parts = cugraph.ecg(G)
parts.to_csv('/content/drive/MyDrive/data/annex_ecg.csv')
```
